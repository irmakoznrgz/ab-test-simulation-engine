```python
import statsmodels.stats.api as sms

base_conversion_rate = 0.05  
mde_relative = 0.20         
target_conversion_rate = base_conversion_rate * (1 + mde_relative) 
alpha = 0.05                 
power = 0.80       
          
effect_size = sms.proportion_effectsize(base_conversion_rate, target_conversion_rate)

required_n = sms.NormalIndPower().solve_power(
    effect_size, 
    power=power, 
    alpha=alpha, 
    ratio=1 
)

print(f"Current Rate: {base_conversion_rate * 100}%")
print(f"Target Rate: {target_conversion_rate * 100}%")
print(f"Minimum number of users required per variation (A and B): {int(required_n)}")
print(f"TOTAL number of users required for the test: {int(required_n) * 2}")
```
